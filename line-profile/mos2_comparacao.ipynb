```python
from matplotlib import pyplot as plt
from skimage.measure import profile_line
from skimage.io import imread
import numpy as np
```

```python
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.size"] = 8
# plt.rcParams["font.serif"] = ["Times"]
```

```python
inches_to_cm = 0.393701
```

```python
one_layer = imread('mos2_one_layer.tif') #lendo a imagem
```

```python
one_layer[46,121]
```

```python
plt.imshow(one_layer==144)
plt.imshow(one_layer, alpha=0.5, cmap="Greys")
```

```python
one_layer.max(), one_layer.min()
```

```python
two_layers = imread('mos2_two_layers.tif')
```

```python
two_layers.max(), two_layers.min()
```

```python
plt.imshow(two_layers==255)
plt.imshow(two_layers, alpha=0.5, cmap="Greys")
```

```python
rows, cols = np.where(one_layer==255)
```

```python
rows, cols
```

```python
start = [116,0]
end = [46,121]
profile_one = profile_line(one_layer, start, end)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20,10))
fig.suptitle('Line Profile Mos2 with one layer')
ax[0].set_title('Image')
ax[0].imshow(one_layer)
ax[0].plot([start[1], end[1]], [start[0], end[0]], 'r')
ax[1].set_title('Profile')
ax[1].plot(profile_one)
```

```python
profile_two = profile_line(two_layers, start, end)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20,10))
fig.suptitle('Line Profile Mos2 with two layers')
ax[0].set_title('Image')
ax[0].imshow(two_layers)
ax[0].plot([start[1], end[1]], [start[0], end[0]], 'r')
ax[1].set_title('Profile')
ax[1].plot(profile_two)
```

```python
plt.figure(figsize=(10,10))
plt.title('Line Profile of MoS2')
plt.plot(profile_one, '--r', label = 'One layer')
plt.plot(profile_two, "-b", label = 'Two layers', alpha=0.5)
plt.legend()
plt.show()
```

```python
three_layers = imread('mos2_three_layers.tif')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20,10))
fig.suptitle('Line Profile Mos2 with three layers')
ax[0].set_title('Image')
ax[0].imshow(three_layers)
ax[0].plot([start[1], end[1]], [start[0], end[0]], 'r')
ax[1].set_title('Profile')
ax[1].plot(profile_three)
```

```python
start = [0,0]
end = 70,121
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20*inches_to_cm,10*inches_to_cm))
# fig.suptitle(')
ax[0].set_title('Imagem Simulada')
ax[0].imshow(one_layer_vs1)
ax[0].plot([start[1], end[1]], [start[0], end[0]], 'r')
ax[1].set_title('Line Profile')
ax[1].plot(profile_line(one_layer_vs1, start, end))
plt.savefig('line-profile-one-layer-vs1.pdf', dpi=900, bbox_inches='tight')
```

```python
two_layers_vs1_1l = imread('mos2_two_layers_vs1_1l.tif')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20*inches_to_cm,10*inches_to_cm))
# fig.suptitle('Line Profile Mos2 with two layers and VS1 vacancy @ 1st layer')
ax[0].set_title('Imagem Simulada')
ax[0].imshow(two_layers_vs1_1l)
ax[0].plot([start[1], end[1]], [start[0], end[0]], 'r')
ax[1].set_title('Line Profile')
ax[1].plot(profile_line(two_layers_vs1_1l, start, end))
plt.savefig('line-profile-two-layers-vs1-1.pdf', dpi=900, bbox_inches='tight')
```

```python
two_layers_vs1_2l = imread('mos2_two_layers_vs1_2l.tif')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20*inches_to_cm,10*inches_to_cm))
# fig.suptitle('Line Profile Mos2 with two layers and VS1 vacancy @ 2nd layer')
ax[0].set_title('Imagem Simulada')
ax[0].imshow(two_layers_vs1_2l)
ax[0].plot([start[1], end[1]], [start[0], end[0]], 'r')
ax[1].set_title('Line Profile')
ax[1].plot(profile_line(two_layers_vs1_2l, start, end))
plt.savefig('line-profile-two-layers-vs1-2.pdf', dpi=900, bbox_inches='tight')
```

```python
plt.figure(figsize=(20*inches_to_cm,20*inches_to_cm))
# plt.title('Line Profile of MoS2: VS1 Comparison')
plt.plot(profile_line(two_layers_vs1_1l, start, end), '--r', label = 'Bicamada com defeito na primeira camada')
plt.plot(profile_line(two_layers_vs1_2l, start, end), '--g', label= 'Bicamada com defeito na segunda camada')
plt.plot(profile_line(one_layer_vs1, start, end), "-b", label = 'Monocamda com defeito', alpha=0.5)
plt.legend(loc='upper left')
plt.savefig('line-profile-comp.pdf', dpi=900, bbox_inches='tight')

plt.show()
```

```python
three_layers_vs1_1l = imread('mos2_three_layers_vs1_1l.tif')
three_layers_vs1_2l = imread('mos2_three_layers_vs1_2l.tif')
three_layers_vs1_3l = imread('mos2_three_layers_vs1_3l.tif')
```

```python
fig, ax = plt.subplots(3, 2, figsize=(12,15))
fig.suptitle('Line Profile Mos2 with three layers and VS1 vacancy @ 2nd layer')
ax[0,0].set_title('VS1 vacancy @ 1st layer')
ax[0,0].imshow(three_layers_vs1_1l)
ax[0,0].plot([start[1], end[1]], [start[0], end[0]], 'r')
ax[1,0].set_title('VS1 vacancy @ 2nd layer')
ax[1,0].imshow(three_layers_vs1_2l)
ax[1,0].plot([start[1], end[1]], [start[0], end[0]], 'r')
ax[2,0].set_title('VS1 vacancy @ 3rd layer')
ax[2,0].imshow(three_layers_vs1_3l)
ax[2,0].plot([start[1], end[1]], [start[0], end[0]], 'r')
ax[0,1].set_title('Profile')
ax[1,1].set_title('Profile')
ax[2,1].set_title('Profile')
ax[0,1].plot(profile_line(three_layers_vs1_1l, start, end))
ax[1,1].plot(profile_line(three_layers_vs1_2l, start, end))
ax[2,1].plot(profile_line(three_layers_vs1_3l, start, end))
```

```python
plt.figure(figsize=(10,10))
plt.title('Line Profile of MoS2 with 3 layers: VS1 Comparison ')
plt.plot(profile_line(three_layers_vs1_1l, start, end), '--r', label = 'Three layers vs1 @ 1st layer')
plt.plot(profile_line(three_layers_vs1_2l, start, end), label= 'Three layers vs1 @ 2nd layer')
plt.plot(profile_line(three_layers_vs1_3l, start, end), "-b", label = 'Three layers vs1 @ 3rd layer', alpha=0.5)
plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.title('Line Profile of MoS2')
plt.plot(profile_line(one_layer, start, end), label = 'One layer')
plt.plot(profile_line(two_layers,start, end), label = 'Two layers')
plt.plot(profile_line(three_layers,start, end), label = 'Three layers')
plt.plot(profile_line(one_layer_vs1, start, end),'--', label = 'One layers with vs1')
plt.plot(profile_line(two_layers_vs1_1l, start, end), '--',label = 'Two layers with vs1 @ 1st layer')
plt.plot(profile_line(two_layers_vs1_2l, start, end),'--', label = 'Two layers with vs1 @ 2nd layer')
plt.plot(profile_line(three_layers_vs1_1l, start, end), label = 'Three layers vs1 @ 1st layer')
plt.plot(profile_line(three_layers_vs1_2l, start, end), label= 'Three layers vs1 @ 2nd layer')
plt.plot(profile_line(three_layers_vs1_3l, start, end), label = 'Three layers vs1 @ 3rd layer')
plt.legend(loc='upper left')
plt.show()
```
